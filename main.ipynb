```python
import pandas as pd
import pyautogui
import time

link = "https://dlp.hashtagtreinamentos.com/python/intensivao/login"

df_produtos = pd.read_csv('produtos.csv')
df_produtos.head()
```

```python
#abrir o google
time.sleep(2)
pyautogui.press('win')
pyautogui.write("Chrome")
time.sleep(0.5)
pyautogui.press("enter")

#abrir o link
time.sleep(2)
pyautogui.write(link)
pyautogui.press("enter")

#entrar no login

time.sleep(5)
pyautogui.press("tab")
pyautogui.write('teste')
time.sleep(2)
pyautogui.press("tab")
pyautogui.write('teste')

time.sleep(1)
pyautogui.press("tab")
pyautogui.press("enter")

time.sleep(2)
pyautogui.press("enter")

#cadastrar produto

for linha in df_produtos.index:
	pyautogui.click(572, 262)
	pyautogui.write(df_produtos.loc[linha, 'codigo'])
	#tab
	pyautogui.press('tab')
	pyautogui.write(str(df_produtos.loc[linha, 'marca']))
	#tab
	pyautogui.press('tab')
	pyautogui.write(str(df_produtos.loc[linha, 'tipo']))
	#tab
	pyautogui.press('tab')
	pyautogui.write(str(df_produtos.loc[linha, 'categoria']))
	#tab
	pyautogui.press('tab')
	pyautogui.write(str(df_produtos.loc[linha, 'preco_unitario']))
	#tab
	pyautogui.press('tab')
	pyautogui.write(str(df_produtos.loc[linha, 'custo']))

	obs = df_produtos.loc[linha, 'obs']
	pyautogui.press('tab')
	if not pd.isna(obs):
		pyautogui.write(str(df_produtos.loc[linha, 'obs']))
	pyautogui.press('tab')
	time.sleep(1)
	for i in range(3):
		pyautogui.press('enter')
	break
	#time.sleep(2)
	
```
